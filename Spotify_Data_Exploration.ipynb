## Data Analysis

### Top 5 most popular Artists

```python
top_five_artists = df.groupby("artist").count().sort_values(by = "song_title", ascending = False)["song_title"][:5]
```

```python
top_five_artists
```

```python
top_five_artists.plot.barh()
plt.xlabel("Number of songs")
plt.ylabel("Artist")
plt.title("Top 5 Artists")
plt.show()
```

### Top 5 loudest tracks

```python
top_five_loudest_tracks = df[['loudness', 'song_title']].sort_values(by="loudness", ascending = True)[:5]
```

```python
top_five_loudest_tracks
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x='loudness', y='song_title', data=top_five_loudest_tracks)
plt.xlabel("Loudness")
plt.ylabel("Track Name")
plt.title("Top 5 loudest tracks")
plt.show()
```

### Artist with the most danceability song

```python
top_five_danceabiliy_song = df[["danceability", "artist", "song_title"]].sort_values(by="danceability", ascending=False)[:5]
top_five_danceabiliy_song
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x='danceability', y='artist', data=top_five_danceabiliy_song)
plt.xlabel("Danceable")
plt.ylabel("Artist Name")
plt.title("Top 5 Artist with danceable tracks")
plt.show()
```

### Top 10 instrumentalness song

```python
top_ten_instrumentalness_song = df[['instrumentalness', 'song_title', 'artist']].sort_values(by='instrumentalness', ascending=False)[:10]
top_ten_instrumentalness_song
```

```python
plt.figure(figsize=(12,7))
plt.pie(x='instrumentalness', data=top_ten_instrumentalness_song, autopct='%1.2f%%', labels=top_ten_instrumentalness_song.song_title)
plt.title("Top 10 song without instruments")
plt.show()
```

### Multiple feature plot

```python
df.columns
```

```python
interest_feature_cols = ["tempo", "loudness", "acousticness", "danceability", "duration_ms", "energy", "instrumentalness", "liveness", "speechiness", "valence"]
```

```python
for i in interest_feature_cols:
    pos_data = df[df["target"] == 1][i]
    neg_data = df[df["target"] == 0][i]
    
    plt.figure(figsize=(12,7))
    
    sns.distplot(pos_data, bins=30, label="Positive", color="green")
    sns.distplot(neg_data, bins=30, label="Negative", color="red")
    
    plt.legend(loc="upper right")
    plt.title(f"Positive and Negative Histogram Plot for {i}")
    plt.show()
```

### Top 10 energetic tracks

```python
top_ten_energetic_song = df[['energy', 'song_title', 'artist']].sort_values(by='energy', ascending=False)[:10]
top_ten_energetic_song
```

```python
plt.figure(figsize=(12,7))
#Making Donut Chart
my_circle = plt.Circle( (0,0), 0.7, color='white')

plt.pie(x='energy', data=top_ten_energetic_song, autopct='%1.2f%%', labels=top_ten_energetic_song.song_title, wedgeprops = { 'linewidth' : 7, 'edgecolor' : 'white' })
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.title("Top 10 energetic song")
plt.show()
```

### Average Duration of tracks

```python
average_duration = df['duration_ms'].mean()
print("Average duration of track is "+str(round(average_duration/1000))+' sec')
```
